Em uma regressão linear simples estamos interessados em desenvolver uma equação linear que apresente a relação entre uma variável dependente (y) e uma variável explicativa (x), sem que, necessariamente, ocorra uma relação de causa e efeito.

<p><font size = 4 face = 'verdana'>1) O conceitos de regressão será apresentado abaixo por meio de um exemplo que considera apenas uma variável explicativa (regressão linear simples). Imagine que, em determinado dia de aula um professor tenha o interesse em saber, para uma turma de 1O estudantes de uma mesma classe, qual a relação entre a distância percorrida para se chegar à escola e o tempo de percurso. Sendo assim, o professor elaborou um questionamento com cada um dos seus 10 alunos e montou um banco de dados que se encontra abaixo:</font></p>

```python
#Importando a bilbioteca Pandas
import pandas as pd
import numpy as np

#Criando o numpy.array (data)
data =np.array([[15, 8],
                [20, 6],
                [20, 15],
                [40, 20],
                [50, 25],
                [25, 11],
                [10, 5],
                [55, 32],
                [35, 28],
                [30, 20]])

#Criando o DataFrame com o np.array (data) com as colunas 'Tempo (y)', 'Distancia (X)'
df = pd.DataFrame(data,columns=['Tempo (y)', 'Distancia (X)'])

# Mostrando o nosso dataset que foi criado
df.head(10)
```

```python
# Passo 1: Calculando a média das colunas Tempo (y) e Distancia (X)
media_tempo = df['Tempo (y)'].mean()
media_distancia = df['Distancia (X)'].mean()

# Imprimindo na tela
print('A média do Tempo (y) é: ', media_tempo)
print('A média do Distancia (X) é: ', media_distancia)
```

```python
# Passo 2: Subtrair os valores observados de tempo e distancia e sua respectiva média
df['Xi-Xmedia'] = df['Distancia (X)'] - media_distancia
df['Yi-Ymedia'] = df['Tempo (y)'] - media_tempo

# Passo 3: Multiplique (Xi-Xmedia) X (Yi-Ymedia)
df['(Xi-Xmedia).(Yi-Ymedia)'] = df['Xi-Xmedia'] * df['Yi-Ymedia']

# Passo 4: Faça (Xi-Xmedia)²
df['(Xi-Xmedia)²'] = df['Xi-Xmedia'] ** 2

# Mostrar o dataset completo com os calculos
df.head()
```

```python
# Passo 5: Some os valores das colunas '(Xi-Xmedia).(Yi-Ymedia)' e '(Xi-Xmedia)²
somatorio_numerador = df['(Xi-Xmedia).(Yi-Ymedia)'].sum()
somatorio_denominador = df['(Xi-Xmedia)²'].sum()

# Imprimindo o resultado na tela
print('Somatório de (Xi-Xmedia).(Yi-Ymedia): ', somatorio_numerador)
print('Somatório de (Xi-Xmedia)²: ', somatorio_denominador)
```

<p><font size: 4 face = 'verdana'>Para calcular os estimadores/ coeficientes alfa e beta, vamos usar as expressões de acordo com Fávero e Belfiore (2021) e os valores calculados no dataset. Vale ressaltar que as equações abaixo seguem 2 critérios:

<li><b>Soma dos erros igual a zero:</b> Com apenas esta primeira condição. podem ser encontradas diversas retas de regressão em que a somatória dos erros (resíduos) seja zero. Portanto, faz-se necessário o estabelecimento de uma segunda condição.</li> 
<li><b>Soma dos erros ao quadrado sendo a mínima possível:</b> Essa condição vai escolher a melhor reta que se ajusta à nuvem de pontos. Logo estamos determinando alfa e beta de modo que a somatória dos quadradrados dos erros (resíduos) seja a menor possível. Assim chamado de método de Mínimos Quadrados Ordinários - MQO, ou em inglês, Ordinary Least Squares - OLS.</li>

A grande pergunta é: Quais valores de alfa e beta que minimizam a soma dos erros², garantindo que a soma dos erros seja? Eis as equações abaixo que garantem essas condições.</font></p>

<img src = "https://github.com/oserxavier/Machine-Learning/blob/main/figures/img_1.png?raw=true" width="200" height="150">

```python
# Passo 6: Com os resultados encontrados, vamos calcular os valores de alfa e beta
beta = somatorio_numerador/somatorio_denominador
alfa = media_tempo - beta * media_distancia

# Imprimindo os valores
print('Valor do intercepto (coeficiente linear/alfa) da reta é: ', round(alfa,4))
print('Valor da inclinação (coeficiente angular/beta) da reta é: ', round(beta,4))
```

```python
# Passo 7: Os critérios foram estão dentro do esperado? Vamos preparar a base de dados

# 1 - Vamos criar o mesmo dataset que criamos originalmente
#Criando o numpy.array (data)
data =np.array([[15, 8],
                [20, 6],
                [20, 15],
                [40, 20],
                [50, 25],
                [25, 11],
                [10, 5],
                [55, 32],
                [35, 28],
                [30, 20]])

#Criando o DataFrame com o np.array (data) com as colunas 'Tempo (y)', 'Distancia (X)'
df = pd.DataFrame(data,columns=['Tempo (y)', 'Distancia (X)'])

# 2 - Calcule o valor esperado de tempo para cada observação considerando o que foi encontrado na equação y = alfa + beta.X
df['tempo_esp'] = 5.8784 + 1.4189 * df['Distancia (X)']

# 3 - Calcule o erro Y (observado) - Y(esperado)
df['erro'] = df['Tempo (y)'] - df['tempo_esp']

# 4. Calcule o erro²
df['erro²'] = df['erro'] **2

# soma dos erros e erro²
soma_erro = df['erro'].sum()
soma_erro_quad = df['erro²'].sum()


print('')
print('Somatório dos erros é: ', round(soma_erro, 2))
print('Somatório dos erros² é: ', round(soma_erro_quad, 2))

df.head(10)
```
